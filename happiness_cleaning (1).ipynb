```python
lists = [df_2015, df_2016, df_2017, df_2018, df_2019]

# Common countries 
common_countries = my_list

def update_dataframe(df, common_countries):
    return df[df['country'].isin(common_countries)]

# Update df with the list
for idx, df in enumerate(lists):
    lists[idx] = update_dataframe(df, common_countries)

# Print df updated
for idx, df in enumerate(lists):
    print(f"DataFrame {idx + 1} après mise à jour :")
    print(df)
    print()
```

```python
filtered_df = df_deaths

# Merge filtered data with corresponding year's data
def merge_data(filtered_df, year_df, year):
    filtered_year_df = filtered_df[filtered_df['year'] == year].drop(columns=['year'])
    merged_df = pd.merge(year_df, filtered_year_df, on='country', how='left')
    return merged_df

# Merge each year's data
df_2015_2 = merge_data(filtered_df, df_2015_1, 2015)
df_2016_2 = merge_data(filtered_df, df_2016_1, 2016)
df_2017_2 = merge_data(filtered_df, df_2017_1, 2017)
df_2018_2 = merge_data(filtered_df, df_2018_1, 2018)
df_2019_2 = merge_data(filtered_df, df_2019_1, 2019)
```

```python
import requests
# Function to get flag URL
def get_flag_url(country):
    response = requests.get(f"https://restcountries.com/v3.1/name/{country}")
    if response.status_code == 200:
        country_info = response.json()[0]
        return country_info['flags']['png']
    else:
        return None

# Add flag URLs to the DataFrame
final_df['Flag_URL'] = final_df['country'].apply(get_flag_url)

# Display the DataFrame with flag URLs
print(final_df)
```
